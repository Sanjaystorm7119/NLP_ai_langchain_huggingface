```python
from nltk.stem import PorterStemmer
from nltk.corpus import words
from random import sample

# random words
random_words = sample(words.words(), 50)
print(f"rand_words : {random_words}")

# initialize stemmer
porter = PorterStemmer()

# stem each word
stemmed_words = [porter.stem(word) for word in random_words]
print(f"stemmed : {stemmed_words}")

# pair original words with stems
pairs = list(zip(random_words, stemmed_words))
print(pairs)
```

```python
#snowball stemmer
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer(language="english")

stemmed = [snowball_stemmer.stem(word) for word in random_words]

# print([word:stem for word , stem in zip(random_words,stemmed)])
pairs = [[word, stem] for word, stem in zip(random_words, stemmed)]

for pair in pairs:
    print(pair)


```
