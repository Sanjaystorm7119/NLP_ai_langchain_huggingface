```python
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk

corpus = """Lorem's ipsum dolor sit amet consectetur adipisicing elit. Dolorum qui unde dolor voluptatibus's doloribus eaque atque hic nesciunt! Soluta ut, incidunt velit sed molestiae vitae nam quo officia laboriosam optio debitis reprehenderit minus corporis quam et, omnis, nulla modi enim consectetur dolorem at. Maxime iusto assumenda quia reprehenderit, illo laboriosam officia quasi quidem quis consequuntur similique possimus, cumque eius, suscipit ut inventore enim ab? Facere magni neque velit necessitatibus iure inventore molestias obcaecati perspiciatis maxime fugiat ratione nam vero magnam voluptatibus, pariatur saepe consequuntur reprehenderit itaque. Quasi in nesciunt at fugiat dignissimos dolor accusamus aliquam laborum alias.Itaque, minus temporibus
"""

documents = sent_tokenize(corpus)
print(documents)
```

```python
#word tokenise
print(corpus)

corpus_word_tokenised = nltk.word_tokenize(corpus)
documents_word_tokenised = [word_tokenize(sentence) for sentence in documents]

print(corpus_word_tokenised)
print(documents_word_tokenised)


```

```python
#wordpunct tokenise

from nltk import  wordpunct_tokenize

PunktTokenizer_word = wordpunct_tokenize(corpus)

print(PunktTokenizer_word)


```

```python
#tree bank word tokenise
from nltk.tokenize import TreebankWordTokenizer

tree_bank_word = TreebankWordTokenizer().tokenize(corpus)
tree_bank_word_sentence = TreebankWordTokenizer()

tree_bank_word_sentence = [TreebankWordTokenizer().tokenize(sentence) for sentence in documents]

print(f"tree_bank_word : {tree_bank_word}",end="\n+++++++++++\n")
print(f"tree_bank_word_sentence: {tree_bank_word_sentence}",end="\n ++++++++ \n")

for sentence in documents:
    print(TreebankWordTokenizer().tokenize(sentence))
```
